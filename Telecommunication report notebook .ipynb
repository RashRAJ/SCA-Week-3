```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
%matplotlib inline
```

```python
df = pd.read_csv('telecom_churn.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
print(df['International plan'].value_counts(normalize=True)*100)
print(df['Voice mail plan'].value_counts(normalize=True)*100)
```

```python
enc = LabelEncoder()
df['International plan'] = enc.fit_transform(df['International plan'])
df['Voice mail plan'] = enc.fit_transform(df['Voice mail plan'])
df['Churn'] = enc.fit_transform(df['Churn'])
```

```python
df.head()
```

```python
df['total minutes per day'] = (df['Total day minutes']+ df['Total eve minutes'] + df['Total night minutes']+ df['Total intl minutes'])
df['total calls per day'] = (df['Total day calls']+ df['Total eve calls'] + df['Total night calls'] + df['Total intl calls'])
df['total charge per day'] = (df['Total day charge']+ df['Total eve charge'] + df['Total night charge'] + df['Total intl charge'])
```

```python
call_states_sum = df.groupby(['State'], as_index=False)['total calls per day'].sum().max()
print(call_states_sum)
charge_states_sum = df.groupby(['State'], as_index=False)['total charge per day'].sum().max()
print(charge_states_sum)
```

```python
df.describe()
```

```python
correlation = df.corr()

f, ax = plt.subplots(figsize=(14,12))
plt.title('Correlation of numerical attributes', size=16)
sns.heatmap(correlation)
plt.show()
```

```python
y = df['Churn']
```

```python
df.drop(['State', 'Churn'], 1, inplace=True)
X = df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_auc_score
import scikitplot as skplt
from sklearn import metrics
```

```python
X_train,X_val, y_train, y_val = train_test_split(X,y, test_size=0.2,  random_state= 0)
```

```python
normaliser = StandardScaler()
X_train = normaliser.fit_transform(X_train)
X_val = normaliser.transform(X_val)
```

```python
clf  = DecisionTreeClassifier(max_depth=10, random_state=100)
clf.fit(X_train, y_train)
score = clf.score(X_val,y_val)
score
```

```python
pred = clf.predict(X_val)
score = roc_auc_score(y_val, pred)
print(score)
```

```python
without my additional features the model score was 0.926 and the accuracy score was 0.866
```

```python
clf.fit(X_train, y_train)
pred = clf.predict(X_val)
skplt.metrics.plot_confusion_matrix(pred, y_val, title="Decesion tree classifier",
            normalize=True,figsize=(6,6),text_fontsize='large')
plt.show()
```
